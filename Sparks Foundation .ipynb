## 5. Team Analysis :-

### 5.1 Teams Performance in all matches played

```python
g_team1 = matches.groupby('team1')
g_team2 = matches.groupby('team2')
g_winner = matches.groupby('winner')
total_matches = pd.DataFrame(data= (g_team1.size()+g_team2.size()), columns=['Total Matches'] )
total_matches['Winning Matches'] = g_winner.size()
total_matches['Total Win by Runs'] = g_winner['win_by_runs'].agg("sum")
total_matches['Total Wickets Saved'] = g_winner['win_by_wickets'].agg("sum")
total_matches['Success Rate'] = (total_matches['Winning Matches']/total_matches['Total Matches'])*100
total_matches.sort_values(by='Winning Matches', ascending=False)[total_matches['Total Matches']>50]
```

### 5.2 Teams Total Bowling Performance 

```python
g_team_balling = deliveries.groupby('bowling_team')
teams_balling_statistics = pd.DataFrame(data={'Runs Given':g_team_balling['total_runs'].agg("sum") })
teams_balling_statistics['Total Balls'] = g_team_balling.size()
teams_balling_statistics['Total Wickets Taken'] = g_team_balling['player_dismissed'].count()
teams_balling_statistics['Total Fielding'] = g_team_balling['fielder'].count()
teams_balling_statistics['Economy'] = teams_balling_statistics['Runs Given']/teams_balling_statistics['Total Balls']
teams_balling_statistics['Out Percent'] = (teams_balling_statistics['Total Wickets Taken']/teams_balling_statistics['Total Balls'])*100
teams_balling_statistics[teams_balling_statistics['Total Balls']>7000].sort_values(by="Out Percent", ascending = False)
```

### 5.3 Teams Bowling Performance Inning wise

```python
balling_first_innings = deliveries[deliveries['inning']==1].groupby('bowling_team')
balling_second_innings = deliveries[deliveries['inning']==2].groupby('bowling_team')
balling_innings = pd.DataFrame(data={'First Inning':(balling_first_innings['total_runs'].agg('sum')/balling_first_innings.size())})
balling_innings['Second Inning'] = balling_second_innings['total_runs'].agg('sum')/balling_second_innings.size()
balling_innings['Balls'] = balling_first_innings.size()+balling_second_innings.size()
balling_innings
```

### 5.4 Teams Total Batting Performance

```python
g_team_batting = deliveries.groupby('batting_team')
teams_batting_statistics = pd.DataFrame(data={'Runs Scored':g_team_batting['total_runs'].agg("sum") })
teams_batting_statistics['Wickets Lost'] = g_team_batting['player_dismissed'].count()
teams_batting_statistics['Balls'] = g_team_batting.size()
teams_batting_statistics['Strike Rate'] = round(teams_batting_statistics['Runs Scored']/teams_batting_statistics['Balls'], 2)
teams_batting_statistics['Dismissal Percent'] = (teams_batting_statistics['Wickets Lost']/teams_batting_statistics['Balls'])*100
teams_batting_statistics[teams_batting_statistics['Balls']>7000].sort_values(by="Strike Rate", ascending = False)
```

### 5.5 Teams Batting Performance Inning Wise

```python
batting_first_innings = deliveries[deliveries['inning']==1].groupby('batting_team')
batting_second_innings = deliveries[deliveries['inning']==2].groupby('batting_team')
batting_innings = pd.DataFrame(data={'First Inning':(batting_first_innings['total_runs'].agg('sum')/batting_first_innings.size())})
batting_innings['Second Inning'] = batting_second_innings['total_runs'].agg('sum')/batting_second_innings.size()
batting_innings['Balls'] = batting_first_innings.size()+batting_second_innings.size()
batting_innings
```

## 7.  Factors Affecting Matches

### 7.1 Toss Factor

```python
g_toss_winner = matches[matches['winner']==matches['toss_winner']].groupby('toss_winner')
new_matches=matches[matches['winner']==matches['toss_winner']]
toss_factor = pd.DataFrame(data={'Toss & Match Winner':g_toss_winner.size()})
toss_factor['Chose Balling'] =new_matches[matches['toss_decision']=='field'].groupby('toss_winner').size()
toss_factor['Chose Batting'] = toss_factor['Toss & Match Winner'] - toss_factor['Chose Balling']
toss_factor['Total Toss Winner'] = matches.groupby('toss_winner').size()
toss_factor['Total Toss Appearance'] = g_team1.size()+g_team2.size()
toss_factor['Toss Effect'] = (toss_factor['Toss & Match Winner']/toss_factor['Total Toss Appearance'])*100
toss_factor['Toss Decision Effect'] = (toss_factor['Toss & Match Winner']/toss_factor['Total Toss Winner'])*100
toss_factor[toss_factor['Total Toss Winner']>30]
```

### 7.1.1 Toss decision's effect on winning match

```python
plt.figure(figsize=(15, 8))
toss_factor['Chose Balling'][toss_factor['Total Toss Winner']>30].plot.line(color='#34ff81', marker = 'o')
toss_factor['Chose Batting'][toss_factor['Total Toss Winner']>30].plot.line(color='#1a74dd', marker='o')
plt.title("Toss Decision's Effect on Winning Match", fontsize=25)
plt.xlabel("Teams in IPL", fontsize=20)
plt.ylabel("Number of Decisions", fontsize=20)
plt.legend()
```

### 7.1.2 Matches Won vs Matches Lost when Toss is Won

```python
plt.figure(figsize=(15, 8))
toss_factor['Total Toss Winner'][toss_factor['Total Toss Winner']>30].plot.bar(color='#ff6384')
toss_factor['Toss & Match Winner'][toss_factor['Total Toss Winner']>30].plot.bar(color='#34ff81')
plt.title("Toss Winner", fontsize=25)
plt.xlabel("Teams in IPL", fontsize=20)
plt.ylabel("Number of Tosses", fontsize=20)
plt.legend()
```

### 7.1.3 Total effect of toss on match

```python
plt.figure(figsize=(15, 8))
toss_factor['Toss Effect'][toss_factor['Total Toss Winner']>30].sort_values(ascending=False).plot.bar(color='#febd58')
plt.title("Toss Effect on Outcome", fontsize=25)
plt.xlabel("Teams in IPL", fontsize=20)
plt.ylabel("Effect in %", fontsize=20)
plt.legend()
```

### 7.1.4 Decision and Strategy Knowlege's effect on game

```python
plt.figure(figsize=(15, 8))
toss_factor['Toss Decision Effect'][toss_factor['Total Toss Winner']>30].plot.line(color='green', marker='o')
plt.title("Decision Accuracy", fontsize=25)
plt.xlabel("Teams in IPL", fontsize=20)
plt.ylabel("Decision Accuracy in %", fontsize=20)
# plt.legend()
```

### 7.2 Home-Away effect

```python
city_wins = matches.groupby(['city', 'winner']).apply(lambda x:x['winner'].count()).reset_index(name='Match Wins')
city_wins_pvt = pd.pivot(city_wins, values='Match Wins', index='city', columns='winner')
city_wins_pvt.replace(np.NaN, 0, inplace=True)
city_wins_pvt
```

### 7.2.1 Teams Performance in Different Cities

```python
plt.figure(figsize=(20, 15))
sns.heatmap(city_wins_pvt, annot=True, fmt='g', cmap='flare')
plt.title("Number of Wins in Different City", fontsize=25)
plt.xlabel("Teams in IPL", fontsize=20)
plt.ylabel("Cities Hosted IPL Matches", fontsize=20)
```
